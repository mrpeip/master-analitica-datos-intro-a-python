```python
import pandas as pd

import json
```

```python
# Read the JSON file
with open(r"C:\github\master-analitica-datos-intro-a-python\M4\Ejercicios\Datos\clientes.json", "r", encoding="utf-8") as f:
    data = json.load(f)

# Convert the 'clientes' list to a DataFrame
clientes_df = pd.DataFrame(data["clientes"])
```

```python
ventas_df = pd.read_csv(r"C:\github\master-analitica-datos-intro-a-python\M4\Ejercicios\Datos\ventas.csv")
```
