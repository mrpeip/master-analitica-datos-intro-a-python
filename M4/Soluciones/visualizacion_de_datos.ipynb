```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

# Configuración general
plt.style.use('default')
sns.set_palette("husl")
```

## Ejercicio 1

```python
# Crea un DataFrame con ventas mensuales
datos = {
    'Mes': ['Ene', 'Feb', 'Mar', 'Abr', 'May', 'Jun'],
    'Ventas': [150, 200, 180, 220, 250, 300],
    'Gastos': [80, 90, 85, 100, 110, 120]
}

df_ventas = pd.DataFrame(datos)

# Tu tarea: Crear un gráfico de línea con matplotlib que muestre Ventas y Gastos por Mes
# Usar diferentes colores y agregar leyenda
```

```python
lienzo, grafico = plt.subplots(figsize=(10, 6))

grafico.plot(df_ventas['Mes'], df_ventas['Ventas'], marker='o', label='Ventas')
grafico.plot(df_ventas['Mes'], df_ventas['Gastos'], marker='s', label='Gastos')

grafico.set_title('Ventas vs Gastos Mensuales')
grafico.set_xlabel('Mes')
grafico.set_ylabel('Dinero (€)')
grafico.legend()
grafico.grid(True, alpha=0.3)

plt.show()
```

## Ejercicio 2

```python
# Dataset de ejemplo
datos_productos = {
    'Producto': ['A', 'B', 'C', 'D', 'E'],
    'Ventas_2023': [450, 320, 580, 210, 390],
    'Ventas_2024': [520, 380, 610, 250, 420]
}

df_productos = pd.DataFrame(datos_productos)
```

```python
# Transformar datos para seaborn
df_melted = df_productos.melt(id_vars=['Producto'], 
                             value_vars=['Ventas_2023', 'Ventas_2024'],
                             var_name='Año', 
                             value_name='Ventas')

plt.figure(figsize=(10, 6))
sns.barplot(data=df_melted, x='Producto', y='Ventas', hue='Año')
plt.title('Comparación de Ventas por Producto: 2023 vs 2024')
plt.ylabel('Ventas ($)')
plt.show()
```

```python
# Crear dataset con múltiples variables
np.random.seed(123)
datos_complejos = {
    'Ingresos': np.random.normal(5000, 1500, 100),
    'Publicidad': np.random.normal(1000, 300, 100),
    'Clientes': np.random.normal(200, 50, 100),
    'Satisfaccion': np.random.uniform(1, 10, 100),
    'Ventas': np.random.normal(10000, 2000, 100)
}

df_correlacion = pd.DataFrame(datos_complejos)

# Crear un heatmap de correlaciones usando seaborn
# Incluir anotaciones con los valores de correlación
```

```python
# Calcular matriz de correlación
corr_matrix = df_correlacion.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, 
            annot=True, 
            cmap='coolwarm', 
            center=0,
            square=True,
            fmt='.2f',
            linewidths=0.5)

plt.title('Matriz de Correlación entre Variables')
plt.tight_layout()
plt.show()
```

## Ejercicio 5

```python
# Crear datos de series temporales
fechas = pd.date_range('2023-01-01', periods=365, freq='D')
datos_temporales = {
    'Fecha': fechas,
    'Ventas': np.random.normal(1000, 200, 365) + np.sin(np.arange(365) * 2 * np.pi / 30) * 100,
    'Temperatura': np.random.normal(25, 5, 365) + np.sin(np.arange(365) * 2 * np.pi / 365) * 10
}

df_temporal = pd.DataFrame(datos_temporales)
df_temporal['Mes'] = df_temporal['Fecha'].dt.month

# 1. Gráfico de línea de Ventas a lo largo del tiempo
# 2. Gráfico de promedio mensual de Ventas
# 3. Relación entre Temperatura y Ventas
```

```python
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# Serie temporal completa
axes[0,0].plot(df_temporal['Fecha'], df_temporal['Ventas'], linewidth=1, alpha=0.7)
axes[0,0].set_title('Ventas Diarias - Serie Temporal')
axes[0,0].tick_params(axis='x', rotation=45)

# Promedio mensual
ventas_mensuales = df_temporal.groupby('Mes')['Ventas'].mean()
axes[0,1].bar(ventas_mensuales.index, ventas_mensuales.values, color='orange', alpha=0.7)
axes[0,1].set_title('Promedio Mensual de Ventas')
axes[0,1].set_xlabel('Mes')

# Scatter plot Temperatura vs Ventas
axes[1,0].scatter(df_temporal['Temperatura'], df_temporal['Ventas'], alpha=0.5, color='green')
axes[1,0].set_title('Temperatura vs Ventas')
axes[1,0].set_xlabel('Temperatura')
axes[1,0].set_ylabel('Ventas')

# Boxplot mensual
sns.boxplot(data=df_temporal, x='Mes', y='Ventas', ax=axes[1,1])
axes[1,1].set_title('Distribución Mensual de Ventas')

plt.tight_layout()
plt.show()
```

## Ejercicio 6

```python
# Crear dataset simulado de e-commerce
np.random.seed(789)
n_registros = 1000

datos_ecommerce = {
    'edad': np.random.randint(18, 70, n_registros),
    'ingresos_anuales': np.random.normal(50000, 20000, n_registros),
    'compras_mensuales': np.random.poisson(5, n_registros),
    'monto_promedio_compra': np.random.normal(150, 50, n_registros),
    'satisfaccion_cliente': np.random.randint(1, 11, n_registros),
    'categoria_favorita': np.random.choice(['Electrónicos', 'Ropa', 'Hogar', 'Deportes'], n_registros),
    'dias_ultima_compra': np.random.exponential(30, n_registros)
}

df_ecommerce = pd.DataFrame(datos_ecommerce)
df_ecommerce['gasto_total'] = df_ecommerce['compras_mensuales'] * df_ecommerce['monto_promedio_compra']


# 1. Perfil demográfico de los clientes
# 2. Análisis de comportamiento de compra
# 3. Relación entre variables
# 4. Segmentación por categoría favorita
# 5. Recomendaciones basadas en los hallazgos
```

```python
# Configurar el estilo
plt.style.use('seaborn-v0_8')

# Crear figura principal
fig = plt.figure(figsize=(20, 15))

# 1. Distribución de Edad e Ingresos
ax1 = plt.subplot2grid((3, 4), (0, 0))
ax2 = plt.subplot2grid((3, 4), (0, 1))

ax1.hist(df_ecommerce['edad'], bins=15, alpha=0.7, color='lightblue', edgecolor='black')
ax1.set_title('Distribución de Edad')
ax1.set_xlabel('Edad')

ax2.hist(df_ecommerce['ingresos_anuales'], bins=20, alpha=0.7, color='lightgreen', edgecolor='black')
ax2.set_title('Distribución de Ingresos')
ax2.set_xlabel('Ingresos Anuales')

# 2. Comportamiento de compra por categoría
ax3 = plt.subplot2grid((3, 4), (0, 2), colspan=2)
sns.boxplot(data=df_ecommerce, x='categoria_favorita', y='gasto_total', ax=ax3)
ax3.set_title('Gasto Total por Categoría Favorita')
ax3.tick_params(axis='x', rotation=45)

# 3. Relación Ingresos vs Gasto
ax4 = plt.subplot2grid((3, 4), (1, 0), colspan=2)
scatter = ax4.scatter(df_ecommerce['ingresos_anuales'], 
                     df_ecommerce['gasto_total'],
                     c=df_ecommerce['satisfaccion_cliente'],
                     cmap='viridis', alpha=0.6)
ax4.set_xlabel('Ingresos Anuales')
ax4.set_ylabel('Gasto Total')
ax4.set_title('Ingresos vs Gasto Total (Color: Satisfacción)')
plt.colorbar(scatter, ax=ax4)

# 4. Heatmap de correlaciones
ax5 = plt.subplot2grid((3, 4), (1, 2), colspan=2)
corr_matrix = df_ecommerce.select_dtypes(include=[np.number]).corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0, ax=ax5)
ax5.set_title('Matriz de Correlación')

# 5. Análisis de frecuencia de compra
ax6 = plt.subplot2grid((3, 4), (2, 0))
compras_por_categoria = df_ecommerce['categoria_favorita'].value_counts()
ax6.pie(compras_por_categoria.values, labels=compras_por_categoria.index, autopct='%1.1f%%')
ax6.set_title('Distribución por Categoría Favorita')

# 6. Satisfacción vs Días desde última compra
ax7 = plt.subplot2grid((3, 4), (2, 1), colspan=3)
sns.scatterplot(data=df_ecommerce, x='dias_ultima_compra', y='satisfaccion_cliente', 
                hue='categoria_favorita', size='gasto_total', alpha=0.7, ax=ax7)
ax7.set_title('Satisfacción vs Tiempo desde Última Compra')
ax7.set_xlabel('Días desde Última Compra')
ax7.set_ylabel('Satisfacción del Cliente')

plt.tight_layout()
plt.show()

# Análisis adicional: Segmentación por gasto
df_ecommerce['segmento_gasto'] = pd.cut(df_ecommerce['gasto_total'], 
                                       bins=3, 
                                       labels=['Bajo', 'Medio', 'Alto'])

print("\n📊 RESUMEN ESTADÍSTICO:")
print(f"Clientes totales: {len(df_ecommerce)}")
print(f"Gasto promedio mensual: ${df_ecommerce['gasto_total'].mean():.2f}")
print(f"Satisfacción promedio: {df_ecommerce['satisfaccion_cliente'].mean():.1f}/10")
print(f"\nCategoría más popular: {df_ecommerce['categoria_favorita'].mode().values[0]}")
```

## Ejercicio 7

```python
np.random.seed(42)
n = 200
datos_3d = {
    'Tamaño': np.random.uniform(10, 50, n),
    'Precio': np.random.uniform(100, 1000, n),
    'Rating': np.random.uniform(1, 5, n),
    'Unidades_Vendidas': np.random.randint(50, 500, n),
    'Categoria': np.random.choice(['A', 'B', 'C'], n),
    'Region': np.random.choice(['Norte', 'Sur', 'Este', 'Oeste'], n)
}

df_3d = pd.DataFrame(datos_3d)

# - Tamaño, Precio y Rating en 3D
# - Color por Categoría
# - Tamaño de marcador por Unidades_Vendidas
# - Información de hover con Region
```

```python
fig = px.scatter_3d(df_3d,
                    x='Tamaño',
                    y='Precio',
                    z='Rating',
                    color='Categoria',
                    size='Unidades_Vendidas',
                    hover_data=['Region'],
                    title='Análisis 3D de Productos: Tamaño vs Precio vs Rating',
                    opacity=0.7)

fig.update_layout(
    scene=dict(
        xaxis_title='Tamaño del Producto',
        yaxis_title='Precio ($)',
        zaxis_title='Rating (1-5)'
    ),
    legend_title='Categoría'
)

fig.show()
```
